# **RESIZE**

```python
# Define transformations
def augment_image(img):
    # Load image
    image = cv2.imread(img)

    # Resize image
    ukuran = (300, 200)
    resized_img = cv2.resize(image, ukuran)

    return resized_img

# Define paths
data_dir = './gdrive/MyDrive/TA/Data/Dataset/eksim'
save_dir = './gdrive/MyDrive/TA/Data/Augmentasi_Data/ASLI/eksim'

# Loop through all images in data_dir
for filename in os.listdir(data_dir):
    # Load image
    img_path = os.path.join(data_dir, filename)
    img = cv2.imread(img_path)

    # Apply transformations
    augmented_image = augment_image(img_path)

    # Save augmented image to save_dir
    save_path = os.path.join(save_dir, filename)
    cv2.imwrite(save_path, augmented_image)

print("done")
```

```python
import random
import shutil
import os

# Define paths
data_dir = './gdrive/MyDrive/TA/Data/Augmentasi_Data/ASLI/eksim'
save_dir = './gdrive/MyDrive/TA/Data/Augmentasi_Data/500data/eksim'

# Get list of all images in data_dir
image_files = os.listdir(data_dir)

# Randomly select 500 images
selected_images = random.sample(image_files, 500)

# Loop through selected images
for filename in selected_images:
    # Copy selected image to save_dir
    src_path = os.path.join(data_dir, filename)
    dst_path = os.path.join(save_dir, filename)
    shutil.copyfile(src_path, dst_path)

print("500 data telah dipilih secara acak dan disimpan di folder tujuan:", save_dir)
```
